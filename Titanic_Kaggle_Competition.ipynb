```python
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
#Import the datasets

train_data = pd.read_csv("train_titanic.csv")
test_data = pd.read_csv("test_titanic.csv")
```

```python
train_data.head()
```

```python
train_data.describe()
```

```python
train_data.isnull().sum()
```

```python
train_x = train_data[["Pclass","Sex"]] #Features
```

```python
train_x.head()
```

```python
#Prediction Target
train_y = train_data[["Survived"]]
train_y.head()
```

```python
# Sklearn Algorithms work with Integers only
#Therefore there is need to convert Sex to integers
#This is done here below!
```

```python
train_x["Sex"].replace("male",1,inplace=True)
train_x["Sex"].replace("female",0,inplace=True)
train_x.head()
```

```python
tr_x,cv_x,tr_y,cv_y = train_test_split(train_x,train_y,test_size=0.30)
```

```python
print(tr_x.head())
```

```python
print(tr_y.head())
```

# Choosing a Training Model

```python
rf = RandomForestClassifier()
```

```python
rf.fit(tr_x,tr_y)
```

# EVALUATING THE MODEL 

```python
accuracy = rf.score(cv_x,cv_y)
print("The Accuracy is:{}%".format(accuracy*100))
```

```python
# Using Logistic Regression algorithm
```

```python
lgr = LogisticRegression()
```

```python
lgr.fit(tr_x,tr_y)
```

```python
accuracy_lgr = lgr.score(cv_x,cv_y)
print("The Accuracy is:{}%".format(accuracy_lgr*100))
```
